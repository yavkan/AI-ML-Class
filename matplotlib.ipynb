```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
mydata = pd.read_csv('mtcars.csv')
```

```python
mydata.rename(columns={'Unnamed: 0':'brand'}, inplace = True)
```

```python
print(mydata.head(10))
```

```python
font = {
    'family' : 'serif',
    'color' : 'black',
    'size' : 12,
    'weight' : 'bold'
}
```

```python
plt.scatter(mydata['mpg'],
           mydata['hp'],
            c = 'DarkBlue',
            s=30)
plt.show()
```

```python
plt.scatter(mydata['mpg'],
           mydata['hp'],
            c = 'DarkBlue',
            s=30)
plt.xlabel('Fuel Efficiency (mpg)',
           fontdict=font,
           labelpad=10)
plt.ylabel('Horse Power',
          fontdict=font,
          labelpad=10)
plt.show()
```

```python
plt.scatter(mydata['mpg'],
           mydata['hp'],
            c = 'DarkBlue',
            s=30)
plt.plot(np.unique(mydata['mpg']),
        np.poly1d(np.polyfit(mydata['mpg'],
                            mydata['hp'], 1))(np.unique(mydata['mpg'])),
        c='DarkBlue'
        )
plt.xlabel('Fuel Efficiency (mpg)',
           fontdict=font,
           labelpad=10)
plt.ylabel('Horse Power',
          fontdict=font,
          labelpad=10)
plt.show()
```
